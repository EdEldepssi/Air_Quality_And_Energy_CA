# County-Level Emissions EDA

### 0. Import Packages

```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Functionalize County-Level EDA

**Functionalizing County Data EDA**

Function should accept a county name, source name, and whether per capita, and produce the corresponding plot:

- Counties:
    - alpine
    - humboldt
    - kern
    - los_angeles
    - marin
    - riverside
    - sacramento
    - san_diego
    - santa_clara
    - yuba
- Sources:
    - eufb (Electric utility fuel burning)
    - cgfb (Cogeneration)
    - ormv (On-road motor vehicles)

```python
def county_EDA(county, source=None, percap=False): # enter county, source as string in above list
    
    # define em_types
    em_types = ['co', 'nh3', 'nox', 'pm', 'pm10', 'pm2_5', 'rog', 'sox', 'tog']
    
    # load appropriate data
    df = pd.read_csv(f'../data/CARB Emissions Data/counties/{county}/{county}_grouping.csv')
    
    # all sources
    if source==None:
        if percap==False:    
            plt.figure(figsize=(15, 9))

            for em in em_types:
                plt.plot(df[df.em_type == em].groupby('year')['tons_per_day'].sum(), label=em)

            plt.xticks(ticks=range(2000, 2022))
            plt.ylabel('Tons per Day', fontdict={'fontsize': 15})
            plt.xlabel('Year', fontdict={'fontsize': 15})
            plt.title(f'{county} Emissions by Type, 2000-21', fontdict={'fontsize': 20})
            plt.legend(fontsize=12);
            
        else:
            plt.figure(figsize=(15, 9))

            for em in em_types:
                plt.plot(df[df.em_type == em].groupby('year')['TpDpCap'].sum(), label=em)

            plt.xticks(ticks=range(2000, 2022))
            plt.ylabel('Tons per Day', fontdict={'fontsize': 15})
            plt.xlabel('Year', fontdict={'fontsize': 15})
            plt.title(f'{county} per Capita Emissions by Type, 2000-21', fontdict={'fontsize': 20})
            plt.legend(fontsize=12);
            
    # source: electric utility fuel burning
    if source=='eufb':
        if percap==False:
            plt.figure(figsize=(15, 9))

            for em in em_types:
                plt.plot(df[(df.em_type == em) & (df.subcategory == 'ELECTRIC UTILITIES')].groupby('year')['tons_per_day'].sum(), label=em)

            plt.xticks(ticks=range(2000, 2022))
            plt.ylabel('Tons per Day', fontdict={'fontsize': 15})
            plt.xlabel('Year', fontdict={'fontsize': 15})
            plt.title(f'{county} Emissions from Electric Utility Fuel Burning by Type, 2000-21', fontdict={'fontsize': 20})
            plt.legend(fontsize=12);

        else:
            plt.figure(figsize=(15, 9))

            for em in em_types:
                plt.plot(df[(df.em_type == em) & (df.subcategory == 'ELECTRIC UTILITIES')].groupby('year')['TpDpCap'].sum(), label=em)

            plt.xticks(ticks=range(2000, 2022))
            plt.ylabel('Tons per Day', fontdict={'fontsize': 15})
            plt.xlabel('Year', fontdict={'fontsize': 15})
            plt.title(f'{county} Emissions per Capita from Electric Utility Fuel Burning by Type, 2000-21', fontdict={'fontsize': 20})
            plt.legend(fontsize=12);
    
    # source: cogeneration
    if source=='cgfb':
        if percap==False:
            plt.figure(figsize=(15, 9))

            for em in em_types:
                plt.plot(df[(df.em_type == em) & (df.subcategory == 'COGENERATION')].groupby('year')['tons_per_day'].sum(), label=em)

            plt.xticks(ticks=range(2000, 2022))
            plt.ylabel('Tons per Day', fontdict={'fontsize': 15})
            plt.xlabel('Year', fontdict={'fontsize': 15})
            plt.title(f'{county} Emissions from Cogeneration Fuel Burning by Type, 2000-21', fontdict={'fontsize': 20})
            plt.legend(fontsize=12);

        else:
            plt.figure(figsize=(15, 9))

            for em in em_types:
                plt.plot(df[(df.em_type == em) & (df.subcategory == 'COGENERATION')].groupby('year')['TpDpCap'].sum(), label=em)

            plt.xticks(ticks=range(2000, 2022))
            plt.ylabel('Tons per Day', fontdict={'fontsize': 15})
            plt.xlabel('Year', fontdict={'fontsize': 15})
            plt.title(f'{county} Emissions per Capita from Cogeneration Fuel Burning by Type, 2000-21', fontdict={'fontsize': 20})
            plt.legend(fontsize=12);
    
    # source: on-road motor vehicles
    if source=='ormv':
        if percap==False:
            plt.figure(figsize=(15, 9))

            for em in em_types:
                plt.plot(df[(df.em_type == em) & (df.category == 'ON-ROAD MOTOR VEHICLES')].groupby('year')['tons_per_day'].sum(), label=em)

            plt.xticks(ticks=range(2000, 2022))
            plt.ylabel('Tons per Day', fontdict={'fontsize': 15})
            plt.xlabel('Year', fontdict={'fontsize': 15})
            plt.title(f'{county} Emissions from On-road Motor Vehicles by Type, 2000-21', fontdict={'fontsize': 20})
            plt.legend(fontsize=12);

        else:
            plt.figure(figsize=(15, 9))

            for em in em_types:
                plt.plot(df[(df.em_type == em) & (df.category == 'ON-ROAD MOTOR VEHICLES')].groupby('year')['TpDpCap'].sum(), label=em)

            plt.xticks(ticks=range(2000, 2022))
            plt.ylabel('Tons per Day', fontdict={'fontsize': 15})
            plt.xlabel('Year', fontdict={'fontsize': 15})
            plt.title(f'{county} Emissions per Capita from On-road Motor Vehicles by Type, 2000-21', fontdict={'fontsize': 20})
            plt.legend(fontsize=12);
```

### 2. Test EDA Function
