```python
df16 = pd.read_excel('../data/wildfires/Fires_100.xlsx', sheet_name='2016', na_values=['<Null>'], 
                   converters= {'ALARM_DATE': pd.to_datetime, 'CONT_DATE': pd.to_datetime})
```

```python
df17 = pd.read_excel('../data/wildfires/Fires_100.xlsx', sheet_name='2017', na_values=['<Null>'], 
                   converters= {'ALARM_DATE': pd.to_datetime, 'CONT_DATE': pd.to_datetime})
print(df17.columns)
print(df16.columns)
```

```python
df17.rename(columns={'YEAR_': 'YEAR'}, inplace=True)
```

```python
df17.columns
```

```python
dffires = pd.concat((df16, df17), axis=0)
dffires.head()
```

```python
dffires['DURATION'] = (dffires['CONT_DATE'] - dffires['ALARM_DATE']).dt.days
```

```python
dffires.info()
```

```python
dffires.dropna(inplace=True)
```

```python
dffires['YEAR'] = dffires['YEAR'].astype(int)
```

```python
dffires = dffires[(dffires.DURATION >= 0) & (dffires.DURATION < 3000)]
dffires.head()
```

```python
dffires.drop(dffires[dffires['STATE'] != 'California'].index, inplace=True)
```

```python
def ex_cause(x):
    try:
        return x.split('_ ')[0].split('- ')[1]
    except:
        return x
```

```python
dffires['CAUSE'] = dffires['CAUSE'].apply(ex_cause)
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
dffires['CAUSE'].value_counts(ascending=False,normalize=True)
```

```python
plt.figure(figsize=(8,6), dpi=200)
dffires['CAUSE'].value_counts(ascending=True,normalize=True).plot(kind='barh', color='#ff9f43');
plt.title('Distribution of wildfire causes from 1940 until 2017')
# plt.savefig('distribution-of-wildfires', bbox_inches='tight')
```

Here we can notice that more since 1940, the amount of wildfires caused by lighning is more than 40% of all fires, baisically it's the main cause of wild fires in California

```python
dffires['YEAR']
```

```python
plt.figure(figsize=(14,8), dpi=120)

sns.histplot(x=dffires['YEAR'].astype(str), hue=dffires['CAUSE'], multiple='stack', bins=87, palette='bright');
plt.xticks(rotation=90, fontsize=7)   
plt.title('Fires and cause over the years')
;
plt.savefig('fires-case-over-time.png', bbox_inches='tight')
```

- We can notice that wildfires were relatively low from 1940 until 2005, then wildfires are immensely spiking up from 2008 until 2017 with lightning being the strongest or cause of fires

```python
df_lightning = dffires[dffires['CAUSE'] == 'Lightning']
```

```python
plt.figure(figsize=(14,8), dpi=120)
sns.histplot(x=df_lightning['YEAR'].astype(str), bins=87, color='#ff9f43')
plt.xticks(rotation=90, fontsize=7);
plt.title('Fires and cause by lightning over the years');
plt.savefig('fires-caused-by-lightning-over-years', bbox_inches='tight')
```

With lightening being the strongest cause of wildfires in califoria, we can notice here that we are getting more lightening starting from 2006 onward

```python
drought = pd.read_csv('../data/drought/dm_export_19800101_20220922.csv', converters={'MapDate': pd.to_datetime})
```

Resources:
    - https://droughtmonitor.unl.edu/DmData/DataDownload/DSCI.aspx

```python
drought['year'] = drought['MapDate'].dt.year
drought.head()
```

```python
drought= drought.drop(columns='Name')
```

```python
drought.groupby('year').mean().head()
```

DSCI: Drought Severity and Coverage Index

```python
plt.figure(figsize=(14,8), dpi=120)

sns.boxplot(x=drought['year'], y= drought['DSCI']);
plt.xticks(rotation=90, fontsize=7)
plt.title('Drought over the years')
;
plt.savefig('drought-over-years', bbox_inches='tight')
```

- With lightening being the strongest cause of the fire, the more drought we have, the more fires tend to spread and stay longer too.

```python
plt.figure(figsize=(8,12), dpi=200)
dffires['UNIT_ID'].value_counts(ascending=True).plot(kind='barh', fontsize=8, color='#ff9f43')
plt.title('Number of wildfires across areas')
#plt.savefig('Number-of-wildfires', bbox_inches='tight')
```

DSCI: Drought Severity and Coverage Index

```python
plt.figure(figsize=(8,8), dpi=100)

dsci_acres = pd.merge(dffires, drought, how='inner', left_on='YEAR', right_on='year')[['CAUSE', 'DSCI']]
dsci_acres.groupby('CAUSE')['DSCI'].mean().plot(kind='bar', color='#ff6b6b')
plt.title('Average Drought correlation to causes of wildfires')
plt.ylabel('Average Drought Per Year');
#plt.savefig('Average-drought-wildfires', bbox_inches='tight')
```

We can see here that drought plays a big role in all kinds of fire in general, specially with causes like railroad fires,
playing with fire, and ofcourse lighning.
