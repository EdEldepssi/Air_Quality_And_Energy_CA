Air pollution of contaminants in our atmosphere. How did these pollutants make their way into our atmosphere? “ The majority of these pollutants are emitted through human activities like burning fossil fuels, vehicle exhaust fumes and emissions from agriculture and industry (ClientEarth). These pollutants enter our body through our respiratory tract. Constant and extended exposure to these contaminants may result in serious health issues. Short term effects include disrupted breathing, asthma, chest pain,and  coughing. And long term effects include heart disease, lung disease, stroke, and even cancer. Currently, there is more research being done on linking pollution to other conditions and diseases as well.  Here are just 4 contaminants in the air and their health effects.

## Particulate matter (PM10, PM2.5)

* nasal and upper respiratory tract health problems
* heart attacks
* strokes
* asthma
* bronchitis
* heart disease 
* cancer 

## Black Carbon


* heart attacks
* stroke
* hypertension
* asthma 
* chronic obstructive pulmonary disease
* bronchitis
* variety of types of cancer

## Nitrogen oxides (NO and NO2)

* asthma
* bronchitis
* higer risk of heart disease

## Ozone (O3)

* chest pain
* coughing
* throat irritation,
* decreased lung function
* chronic obstructive pulmonary disease

As we can see, most of the health related issues from air pollution are heart and lung related or some type of cancer. Let's take a look at mortality rate form various heart/lung related diseases and cancer per county in California.

```python
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel('../data/calenviroscreen40resultsdatadictionary_F_2021.xlsx')
```

```python
df.head(3)
```

```python
df.columns
```

```python
fig = plt.figure(figsize=(8, 9))
ax = fig.add_axes([0,0,1,1])
df = df.sort_values(by='Poverty', ascending=False)
ax.barh(df['California County'],df['Poverty'])
plt.title("Percent of population living below two times the federal poverty level per County ")
plt.xlabel("Percent of population living below two times the federal poverty")
plt.ylabel("County")
plt.show();
```

```python
fig = plt.figure(figsize=(8, 9))
ax = fig.add_axes([0,0,1,1])
df = df.sort_values(by='CES 4.0 Score', ascending=False)
ax.barh(df['California County'],df['CES 4.0 Score'])
plt.title("CalEnviroScreen Score, Pollution Score multiplied by Population Characteristics Score per County ")
plt.xlabel("CalEnviroScreen Score, Pollution Score multiplied by Population Characteristics Score")
plt.ylabel("County")
plt.show();
```

```python
from scipy.stats import pearsonr
sns.set(style='white', font_scale=1.2)

g = sns.JointGrid(data=df, x='CES 4.0 Score', y='Poverty', height=5)
g = g.plot_joint(sns.regplot, color="xkcd:muted blue")
g = g.plot_marginals(sns.distplot, kde=False, bins=12, color="xkcd:bluey grey")
plt.title("Poverty and Pollution Correlation")
plt.tight_layout()
```

```python
sns.set(style='white', font_scale=1.2)

g = sns.JointGrid(data=df, x='Poverty', y='PM2.5 Pctl', height=5)
g = g.plot_joint(sns.regplot, color="xkcd:muted blue")
g = g.plot_marginals(sns.distplot, kde=False, bins=12, color="xkcd:bluey grey")
plt.title("Poverty Score and PM2.5")
plt.tight_layout();
```

```python
sns.set(style='white', font_scale=1.2)

g = sns.JointGrid(data=df, x='CES 4.0 Score', y='Cardiovascular Disease', height=5)
g = g.plot_joint(sns.regplot, color="xkcd:muted blue")
g = g.plot_marginals(sns.distplot, kde=False, bins=12, color="xkcd:bluey grey")
plt.title("Pollution and Cardiovascular Disease")
plt.tight_layout()
```

```python
import statsmodels.api as sm
CES_mean_value = df['CES 4.0 Score'].mean()
Cardio_mean_value = df['Cardiovascular Disease'].mean()
df['CES 4.0 Score'].fillna(value=CES_mean_value, inplace=True)
df['Cardiovascular Disease'].fillna(value=Cardio_mean_value, inplace=True)
x = df['CES 4.0 Score']
y = df['Cardiovascular Disease']


x = sm.add_constant(x)
 
model = sm.OLS(y, x).fit()
predictions = model.predict(x) 
 
print_model = model.summary()
print(print_model)
```
